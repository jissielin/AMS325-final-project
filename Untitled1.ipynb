```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA

# Load primary dataset (Uber stock data)
file_path = 'UBER.csv'  
uber_data = pd.read_csv(file_path)

# Data preprocessing
uber_data['Date'] = pd.to_datetime(uber_data['Date'])  # Convert Date to datetime
uber_data = uber_data.sort_values(by='Date')  # Sort by date

# Load unemployment data (uploaded dataset)
unemployment_data_path = 'USUnemployment.csv'  
us_unemployment_data = pd.read_csv(unemployment_data_path)

# Transform unemployment data from wide to long format
us_unemployment_data = us_unemployment_data.melt(id_vars=['Year'], var_name='Month', value_name='Unemployment Rate')
month_map = {
    'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6,
    'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
}
us_unemployment_data['Month'] = us_unemployment_data['Month'].map(month_map)
us_unemployment_data['Date'] = pd.to_datetime(us_unemployment_data[['Year', 'Month']].assign(Day=1))
us_unemployment_data = us_unemployment_data[['Date', 'Unemployment Rate']]

# Filter unemployment data to match Uber data timeframe
us_unemployment_data = us_unemployment_data[us_unemployment_data['Date'] >= uber_data['Date'].min()]

# Merge datasets on Date
merged_data = pd.merge(uber_data, us_unemployment_data, on='Date', how='inner')

# Summary statistics for merged data
summary_stats = merged_data.describe()
print("Summary Statistics for Merged Data:\n", summary_stats)

# Plotting trends for all datasets
plt.figure(figsize=(12, 6))
plt.plot(merged_data['Date'], merged_data['Close'], label='Uber Closing Price', linewidth=2)
plt.plot(merged_data['Date'], merged_data['Unemployment Rate'], label='Unemployment Rate', linestyle='--', linewidth=2)
plt.title("Trends in Uber Stock and Unemployment Rates", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Values", fontsize=12)
plt.legend(fontsize=10)
plt.grid()
plt.show()

# Correlation analysis
correlation_matrix = merged_data[['Close', 'Volume', 'Unemployment Rate']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix for Merged Data")
plt.show()

# Predictive modeling using ARIMA (example for closing price)
# Prepare data for ARIMA
closing_prices = merged_data['Close']
closing_prices.index = merged_data['Date']

# Train ARIMA model
model = ARIMA(closing_prices, order=(5, 1, 0))  # Example order (p, d, q)
model_fit = model.fit()

# Print model summary
print(model_fit.summary())

# Forecast future prices
forecast = model_fit.forecast(steps=30)  # Predict next 30 days

# Plot forecast
plt.figure(figsize=(12, 6))
plt.plot(closing_prices, label='Historical Data')
plt.plot(forecast, label='Forecast', linestyle='--')
plt.title("ARIMA Forecast of Uber Closing Prices")
plt.xlabel("Date", fontsize=12)
plt.ylabel("Closing Price", fontsize=12)
plt.legend(fontsize=10)
plt.grid()
plt.show()
```
